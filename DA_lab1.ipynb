## EX. 5

```python
sigma = np.absolute(np.random.standard_normal(1000))

x = np.random.normal(0,3,1000)

beta = np.random.normal(0.2,0.2,1000)

alfa = np.random.normal(2,0.1,1000)

u = alfa + beta*x

y = np.random.normal(u,sigma,1000)
```

```python
plt.hist(y,100);
```
